```python
itog_lst = []
import csv

# Open tsv and txt files(open txt file in write mode)
tsv_file = open("/content/audio.tsv")
txt_file = open("7.txt", "w+")

# Read tsv file and use delimiter as \t. csv.reader
# function returns a iterator
# which is stored in read_csv
read_tsv = csv.reader(tsv_file, delimiter="\t")

# write data in txt file line by line
for row in read_tsv:
    if len(row) > 0:
        itog_lst.append({
            "start": int(row[0]) / 1000,
            "end": int(row[1]) / 1000,
            "text": row[-1]
        })

print(itog_lst)
```

```python
for sub in itog_lst:
  print(f"Текст: {sub['text']}\n Время начала: {sub['start']}\n Время окончания: {sub['end']}\n")
```

```python
my_file = open(f"{upload_file[:-4]}.txt", "w+")
for sub in itog_lst:
  my_file.write(f"Текст: {sub['text']}\n Время начала: {sub['start']}\n Время окончания: {sub['end']}\n")
```

```python
text = '' # тут мы переводим


for sub in itog_lst:
    word = sub['text']
    sub['text'] = ts.translate_text(word, translator='google', to_language=prem_lang)
    text += sub['text'] + " "
    my_file.write(f"Текст: {sub['text']}\n Время начала: {sub['start']}\n Время окончания: {sub['end']}\n")

my_file.close()
```

#Запускаем Text2Speash



```python

nltk.download('punkt')

def text_to_audio(text, start, end, language):

    # myobj = gTTS(text=text, lang=language, slow=False)
    audioclip = AudioSegment.from_file("out_audio.wav")
    audioclip = audioclip[start * 1000: end * 1000]
    audioclip.export("prom_say.wav", format="wav")

    tts.tts_to_file(text=text, speaker_wav="prom_say.wav", language=language, file_path="newspeak.wav", speed=4)

    AudioSegment.from_wav("newspeak.wav").export("newspeak.mp3", format="mp3")


all_sound = AudioSegment.from_file("out_audio.mp3", format="mp3")

for i in itog_lst:
  text_to_audio(i["text"], i["start"], i["end"], prem_lang.lower()) #en_speaker_3_long_wav файл с нашим текстом


  new_replic = AudioSegment.from_file("newspeak.mp3", format="mp3")

  start = i["start"]
  end = i["end"]

  new_replic_len = new_replic.duration_seconds
  old_replic_len = end - start

  if old_replic_len < new_replic_len:
    #то ускоряем
    min_len = 0
    x0 = 1
    x = 2
    lr = 0.01
    accelerated_audio = new_replic.speedup(playback_speed=x0)
    while abs(accelerated_audio.duration_seconds - old_replic_len) >= 0.1 and x0 < x:
        accelerated_audio = new_replic.speedup(playback_speed=x0)
        x0 += lr
    accelerated_audio.export("itog.mp3", format="mp3")

  elif old_replic_len > new_replic_len:

    x0 = 1
    x = 0.5
    lr = 0.01
    current_audio_slow_down = ae.speed_down(new_replic, x0)
    while old_replic_len - current_audio_slow_down.duration_seconds >= -0.1:
      current_audio_slow_down = ae.speed_down(new_replic, x0)
      x0 -= lr
    current_audio_slow_down.export("itog.mp3", format="mp3")
  else:
    new_replic.export("itog.mp3", format="mp3")

  itog_new_sound = AudioSegment.from_file("itog.mp3", format="mp3")
  itog_new_sound_len = itog_new_sound.duration_seconds

  start_millisecond = int(round(start * 1000, 0))
  end_millisecond = int(round((start + itog_new_sound_len) * 1000, 0))

  sound_with_new_comment = all_sound[:start_millisecond] + itog_new_sound + all_sound[end_millisecond:]

  all_sound = sound_with_new_comment





music_sound = AudioSegment.from_file("/content/abcd/out_audio/accompaniment.mp3", format="mp3")
output = all_sound.overlay(music_sound, position=0)

output.export("all_sound.mp3", format="mp3")
```
